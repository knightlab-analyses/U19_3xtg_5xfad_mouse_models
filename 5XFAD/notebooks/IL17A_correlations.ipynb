```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import spearmanr
from statannotations.Annotator import Annotator
%matplotlib inline
```

```python
obsfeat = pd.read_csv('../data/Microbiome/all_1000000rar_core_diversity/observed_features_vector/alpha-diversity.tsv', sep='\t', index_col=0)
shannon = pd.read_csv('../data/Microbiome/all_1000000rar_core_diversity/shannon_vector/alpha-diversity.tsv', sep='\t', index_col=0)
faiths = pd.read_csv('../data/Microbiome/all_1000000rar_core_diversity/faith_pd_vector/alpha-diversity.tsv', sep='\t', index_col=0)
```

```python
LI_md = common[common['Tissue']=='LI'].copy()
SI_md = common[common['Tissue']=='SI'].copy()
SP_md = common[common['Tissue']=='SP'].copy()
```

```python
#strongest IL17 signal from SI, second from LI, low in spleen
```

## Observed Features

```python
LI_obsfeat = pd.merge(obsfeat, LI_md, left_index=True, right_index=True)
SI_obsfeat = pd.merge(obsfeat, SI_md, left_index=True, right_index=True)
SP_obsfeat = pd.merge(obsfeat, SP_md, left_index=True, right_index=True)
```

```python
for df in [LI_obsfeat, SI_obsfeat, SP_obsfeat]:
    sns.regplot(data=df, y='observed_features', x='IL17A',
               scatter_kws = {"color": "black", "alpha": 0.5},
                line_kws = {"color": "red"}, ci=99)
    plt.title(df.Tissue.unique(), fontsize=14)
    r, p = spearmanr(df["observed_features"], df["IL17A"])
    plt.show()
    print("spearman", df.Tissue.unique(), "statistic", r, "p-value", p)
    
#SP/spleen has significant correlation, but mostly a straight line
#positive slope, increasing number of detected bacteria, increasing IL17A
```

## Faith PD

```python
LI_faiths = pd.merge(faiths, LI_md, left_index=True, right_index=True)
SI_faiths = pd.merge(faiths, SI_md, left_index=True, right_index=True)
SP_faiths = pd.merge(faiths, SP_md, left_index=True, right_index=True)
```

```python
LI_faiths.head()
```

```python
for df in [LI_faiths, SI_faiths, SP_faiths]:
    sns.regplot(data=df, y='faith_pd', x='IL17A',
               scatter_kws = {"color": "black", "alpha": 0.5},
                line_kws = {"color": "red"}, ci=99)
    plt.title(df.Tissue.unique(), fontsize=14)
    r, p = spearmanr(df["faith_pd"], df["IL17A"])
    plt.show()
    print("spearman", df.Tissue.unique(), "statistic", r, "p-value", p)
    
#SP/spleen has significant correlation, but mostly a straight line
#positive slope, increasing community diversity, increasing IL17A
#"a higher Faith's number means more branches, means more richness (more diversity)"
```

## Shannon

```python
LI_shannon = pd.merge(shannon, LI_md, left_index=True, right_index=True)
SI_shannon = pd.merge(shannon, SI_md, left_index=True, right_index=True)
SP_shannon = pd.merge(shannon, SP_md, left_index=True, right_index=True)
```

```python
LI_shannon.head()
```

```python
for df in [LI_shannon, SI_shannon, SP_shannon]:
    sns.regplot(data=df, y='shannon_entropy', x='IL17A',
               scatter_kws = {"color": "black", "alpha": 0.5},
                line_kws = {"color": "red"}, ci=99)
    plt.title(df.Tissue.unique(), fontsize=14)
    r, p = spearmanr(df["shannon_entropy"], df["IL17A"])
    plt.show()
    print("spearman", df.Tissue.unique(), "statistic", r, "p-value", p)
    
#SI/small intestine and SP/spleen has significant correlation, but mostly a straight line
#positive slope, increasing shannon entropy, increasing IL17A
```

```python
#from John Bostick - bacteria of possible interest
#https://doi.org/10.1038/nature13279
#https://doi.org/10.1073/pnas.1617460113 - Fig 1A
#Look at correlations with Small Intestine primarily; higher abundance 
#What John wants - https://doi.org/10.1002/mds.29300  (Fig 2)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
from statsmodels.stats.multitest import fdrcorrection
from statsmodels.stats.multitest import multipletests
from statannotations.Annotator import Annotator
%matplotlib inline
```

```python
IL17A_SI = pd.read_csv('../data/IL17A_SI_metadata.txt', sep='\t', index_col=0)
IL17A_SI.head()
#sacrifice timepoints: 5 and 8 months only
#Small Intestine Only - most significant findings
```

```python
#RARIFIED
rel_abun_all = pd.read_csv('../data/Microbiome/rar1000000_rel_freq_table.tsv', 
                           skiprows=1, sep='\t', index_col=0)
rar_freq = rel_abun_all.T.copy()
rar_freq.head()
```

```python
rar_freq.shape
```

```python
rar_freq2 = rar_freq.loc[:, rar_freq.sum(axis=0) >= 0.001]
rar_freq2.shape
#removes low frequency microbes
```

```python
bact_list = list(rar_freq2.columns)
```

```python
corr_table = pd.merge(IL17A_SI, rar_freq2, left_index=True, right_index=True)
corr_table
```

```python
bact = []
pval = []
stat = []
for b in bact_list:
    bact.append(b)
    r, p = spearmanr(corr_table['IL17A'], corr_table[b])
    pval.append(p)
    stat.append(r)
```

```python
spear_corr = pd.DataFrame()
spear_corr['genome_id'] = bact
spear_corr['spearman_statistic'] = stat
spear_corr['spearman_pval'] = pval
spear_corr.dropna(subset=['spearman_pval'], inplace=True) #gets rid of static ones with no result
qvals = fdrcorrection(spear_corr['spearman_pval'])[1]
spear_corr['spearman_fdr_qval'] = qvals
spear_corr['spearman_holm_qval'] = multipletests(spear_corr['spearman_pval'], alpha=0.05, method='holm')[1]
spear_corr['IL17A_fdr_qcorrelated'] = spear_corr['spearman_fdr_qval']<0.05
spear_corr['IL17A_holm_qcorrelated'] = spear_corr['spearman_holm_qval']<0.05
spear_corr['IL17A_pcorrelated'] = spear_corr['spearman_pval']<0.05
```
