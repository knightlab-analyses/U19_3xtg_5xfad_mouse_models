```python
microbiome = pd.read_csv('../data/Microbiome/sacrifice/sacrifice_feature_table.tsv', 
                         sep='\t', skiprows=1, index_col=0)
microbiome.head()
#sacrifice only, not longitudinal
```

```python
microbiome.shape
```

```python
metadata = pd.read_csv('../data/Microbiome/14957_metadata.txt', sep='\t',
                      usecols=['sample_name', 'mouse_strain', 'sex', 'host_subject_id', 'host_age'])
metadata.tail()
#rename index from sample tube id to microbiome sample ids
```

```python
metadata.shape
```

```python
metadata['mouse_id'] = metadata['sample_name'].str.split('.').str[1]
metadata.tail()
```

```python
metadata['host_age'] = metadata['host_age'].astype(float)
metadata['Strain'] = metadata['mouse_strain'].str.split(" ", n=1).str[0]
metadata['Strain'] = metadata['Strain'].str.upper()
metadata['study_strain_sex_cage_mouse_tp'] = "5xFAD_SPF_"+metadata['Strain'].astype(str)+"_"+metadata['sex'].astype(str)+"_"+metadata['host_subject_id'].astype(str)+"_"+metadata['mouse_id'].astype(str)+"_"+metadata['host_age'].astype(str)
metadata['study_strain_sex_cage_mouse_tp'].replace(' ', '', regex=True, inplace=True)
metadata.tail()
```

```python
rn_micro = pd.merge(metadata[['sample_name', 'study_strain_sex_cage_mouse_tp']], microbiome.T, 
                    left_on='sample_name', right_index=True)
rn_micro.head()
```

```python
rn_micro.set_index('study_strain_sex_cage_mouse_tp', inplace=True)
rn_micro.head()
```

```python
rn_micro2 = rn_micro.drop(columns=['sample_name'])
rn_micro2.index.name='#OTU ID'
rn_micro2.head()
```

```python
rn_micro2.T.shape
```

```python
rn_micro3 = rn_micro2.T.copy()
rn_micro4 = rn_micro3.groupby(by=rn_micro3.columns, axis=1).mean()
rn_micro4.head()
#duplicate samples, mean
```

```python
rn_micro4.shape
#no duplicates
```

```python
rn_micro4.to_csv('../data/Microbiome/sacrifice/sacrifice_5XFAD_reformat_mouse.tsv', sep='\t')
```

```python
import qiime2 as q2
from qiime2.plugins.emperor.actions import biplot
from qiime2.plugins.gemelli.actions import (joint_rpca, filter_ordination, feature_correlation_table)
from qiime2.plugins.longitudinal.actions import volatility
from qiime2.plugins.diversity.visualizers import beta_group_significance
```

```python
# import tables
microbe = q2.Artifact.load('../data/Microbiome/sacrifice/sacrifice_5XFAD_reformat_mouse.qza')
brain = q2.Artifact.load('../../Multiomics/data/Metabolome/04062024/rawdata_brain_reformat_mouse.qza')
fecal = q2.Artifact.load('../../Multiomics/data/Metabolome/04062024/rawdata_fecal_5xfad_reformat_mouse.qza')
cecal = q2.Artifact.load('../../Multiomics/data/Metabolome/04062024/rawdata_cecal_reformat_mouse.qza')
liver = q2.Artifact.load('../../Multiomics/data/Metabolome/04062024/rawdata_liver_reformat_mouse.qza')
serum = q2.Artifact.load('../../Multiomics/data/Metabolome/04062024/rawdata_serum_reformat_mouse.qza')
# import metadata
metadata = q2.Metadata.load('../data/Metabolome/all_jointRPCA/rpca_metadata_mouse.tsv')
```

```python
# run joint RPCA
jrpca_results = joint_rpca([microbe, brain, fecal, cecal, liver, serum],
                           sample_metadata=metadata,
                           min_feature_frequency=1,
                           max_iterations=3)
jrpca_results.biplot.save('../data/Metabolome/all_jointRPCA/all-micro-metab-bymouse_biplot-ordination.qza')
jrpca_results.distance_matrix.save('../data/Metabolome/all_jointRPCA/all-micro-metab-bymouse_distance-matrix.qza')
jrpca_results.cross_validation_error.save('../data/Metabolome/all_jointRPCA/all-micro-metab-bymouse_cross-validation-error.qza')

jrpca_results
```
