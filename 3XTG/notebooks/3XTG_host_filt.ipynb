```python
import pandas as pd
```

```python
no_host_filt = pd.read_csv('../data/Microbiome/3XTG_no_host_filt_counts.csv', index_col=0)
host_filt = pd.read_csv('../data/Microbiome/3XTG_host_filt_counts.csv', index_col=0)
```

```python
no_host_filt.head()
```

```python
comp = pd.merge(no_host_filt, host_filt, left_index=True, right_index=True)
comp.head()
```

```python
comp.shape
```

```python
comp['differences'] = comp['old_frequency']-comp['new_frequency']
```

```python
comp.head()
```

```python
comp['percent_old'] = comp['differences']/comp['old_frequency']
```
