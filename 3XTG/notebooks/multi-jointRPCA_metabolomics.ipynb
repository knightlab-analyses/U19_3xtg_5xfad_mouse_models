```python
import qiime2 as q2
from qiime2.plugins.emperor.actions import biplot
from qiime2.plugins.gemelli.actions import (joint_rpca, filter_ordination, feature_correlation_table)
from qiime2.plugins.longitudinal.actions import volatility
from qiime2.plugins.diversity.visualizers import beta_group_significance
```

```python
# import tables
brain = q2.Artifact.load('../data/Metabolome/04062024/rawdata_brain_reformat.qza')
fecal = q2.Artifact.load('../data/Metabolome/04062024/rawdata_fecal_reformat.qza')
cecal = q2.Artifact.load('../data/Metabolome/04062024/rawdata_cecal_reformat.qza')
liver = q2.Artifact.load('../data/Metabolome/04062024/rawdata_liver_reformat.qza')
serum = q2.Artifact.load('../data/Metabolome/04062024/rawdata_serum_reformat.qza')
# import metadata
metadata = q2.Metadata.load('../data/Metabolome/04062024/rpca_metadata.tsv')
```

```python
# run joint RPCA
jrpca_results = joint_rpca([brain, fecal, cecal, liver, serum],
                           sample_metadata=metadata,
                           min_feature_frequency=10,
                           max_iterations=6)
jrpca_results.biplot.save('../data/Metabolome/04062024/five-metab_joint_biplot.qza')
jrpca_results.distance_matrix.save('../data/Metabolome/04062024/five-metab_joint_distance_matrix.qza')
jrpca_results.cross_validation_error.save('../data/Metabolome/04062024/five-metab_cross_validation_error.qza')

jrpca_results
```

```python
volatility_plot = volatility(jrpca_results.cross_validation_error.view(q2.Metadata), 'iteration')
volatility_plot.visualization.save('../data/Metabolome/04062024/five-metab_volatility_plot.qzv')
```
