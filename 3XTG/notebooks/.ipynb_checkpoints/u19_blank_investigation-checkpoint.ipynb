```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
plate_df = pd.read_csv('../data/blank_investigation/2023_05_02_AGMP_3XTG_P3_U19_1-3_Celeste_Marm_1_iSeq_pooling_df.txt', sep='\t', index_col=0)
plate_df
```

```python
u19_metadata = pd.read_csv('../data/blank_investigation/14748_20230223-185710.csv')
```

```python
u19_prep = pd.read_csv('../data/blank_investigation/14748_prep_14580_20230404-143720.txt',sep='\t')
```

```python
u19_df_2 = pd.read_csv('../data/blank_investigation/2022_12_28_AGMP_P3_3XTG_U19_4-5_df.txt',sep='\t')
```

```python
u19_df = pd.concat([plate_df,u19_df_2])
```

```python
u19_meta_prep = u19_metadata.merge(u19_prep,on='sample_name',how='left')
```

```python
u19_meta_prep_df = u19_meta_prep.merge(u19_df,left_on='tube_id',right_on='Sample',how='left')
```

```python
u19_meta_prep_df
```

```python
sns.stripplot(x='Blank',y='new_pooling_scaled',data=u19_meta_prep_df)
```

```python
sns.boxplot(x='Blank',y='quality_filtered_reads',data=u19_meta_prep_df,fliersize=0)
sns.stripplot(x='Blank',y='quality_filtered_reads',data=u19_meta_prep_df,color='black',alpha=0.5)
```

```python
u19_meta_prep_df['counts/pooled'] = u19_meta_prep_df['quality_filtered_reads']/u19_meta_prep_df['new_pooling_scaled']
```

```python
sns.boxplot(x='Blank',y='counts/pooled',data=u19_meta_prep_df,fliersize=0)
sns.stripplot(x='Blank',y='counts/pooled',data=u19_meta_prep_df,color='black',alpha=0.5)
plt.yscale('log')
```

```python
u19_ordered_cp_ratio = u19_meta_prep_df.sort_values(by='counts/pooled',ascending=True)
u19_ordered_cp_ratio.head()
```

```python
u19_ordered_cp_ratio[u19_ordered_cp_ratio.sample_name.str.contains("BLANK")]
#set threshold at 2000 quality_filtered_reads per new_pooling_scaled
```

```python
u19_ordered_cp_ratio[u19_ordered_cp_ratio['counts/pooled']<2000].shape
```

```python
u19_ordered_cp_ratio_removal = u19_ordered_cp_ratio[u19_ordered_cp_ratio['counts/pooled']<2000]
u19_ordered_cp_ratio_removal.tube_id.values
```

```python
u19_ordered_cp_ratio_removal.shape
```

```python
u19_ordered_cp_ratio_removal[u19_ordered_cp_ratio_removal.tube_id.str.contains("BLANK")].shape
```

```python
u19_metadata['keep'] = np.where(u19_metadata["tube_id"].isin(u19_ordered_cp_ratio_removal.tube_id.values), "False", "True")
u19_metadata
```

```python
u19_metadata.to_csv('../data/Microbiome/14748_20230223-185710_v2.txt', sep='\t', index=False)
```
