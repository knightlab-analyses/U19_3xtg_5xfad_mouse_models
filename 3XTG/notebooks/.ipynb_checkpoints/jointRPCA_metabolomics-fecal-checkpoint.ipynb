```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Reformat Metabolome file to BIOM

```python
metabolome = pd.read_csv('../data/Metabolome/data_sacrifice_3xtg.csv', index_col=0)
metabolome.index = metabolome.index.map(str)
metabolome.head()
#raw absorbances, no annotations, sacrifice only, not longitudinal
```

```python
metabolome.shape
```

```python
metadata = pd.read_csv('../data/Microbiome/14748_metadata_overpooling.txt', sep='\t',
                      usecols=['sample_name', 'tube_id'], index_col=1)
metadata.tail()
#rename index from sample tube id to microbiome sample ids
```

```python
metadata.shape
```

```python
rn_metab = pd.merge(metadata, metabolome, left_index=True, right_index=True)
rn_metab.head()
```

```python
rn_metab.set_index('sample_name', inplace=True)
rn_metab.head()
```

```python
rn_metab2 = rn_metab.T.copy()
rn_metab2.index.name='#OTU ID'
rn_metab2.head()
```

```python
rn_metab2.to_csv('../data/Metabolome/raw_data_sacrifice_3xtg_reformat.tsv', sep='\t')
```

# Plotting/Graphs

```python
correlation_table = pd.read_csv('../data/Metabolome/sac_jointRPCA/correlation_table/Correlation.tsv',
                               sep='\t', index_col=0)
correlation_table.head()
```

```python
sns.clustermap(correlation_table,
               yticklabels=False,
               xticklabels=False)
plt.show()
```

### Annotations

```python
taxonomy = pd.read_csv('../data/Microbiome/pooling_filtered/sacrifice/RPCA/sac-phyloRPCA-taxonomy/taxonomy.tsv',
                        sep='\t', index_col=0)
taxonomy.head()
```

```python
annot = pd.read_csv('../data/Metabolome/3XTG_fbmn.tsv', sep='\t', 
                    usecols=['#Scan#', 'Compound_Name'], index_col=1)
annot2 = annot.rename('X{}'.format)
annot2.head()
```

```python
renamed = pd.merge(ctable3.T, annot2, left_index=True, right_index=True)
renamed.set_index('Compound_Name', inplace=True)
renamed2 = renamed.T.copy()
renamed2
```

```python
renamed3 = pd.merge(renamed2, taxonomy, left_index=True, right_index=True)
renamed4 = renamed3.set_index('Taxon')
renamed4.head()
```

```python
sns.clustermap(renamed4.apply(pd.to_numeric), figsize=(40, 10))
plt.rcParams['svg.fonttype'] = 'none'
plt.savefig('../data/Metabolome/sac_jointRPCA/VIPgenotype_microtopbot1perc-metab_corr_named.svg', 
           dpi=150, bbox_inches='tight')
```

```python
sns.clustermap(renamed4.apply(pd.to_numeric), z_score=0, figsize=(40, 10))
plt.show()
```
