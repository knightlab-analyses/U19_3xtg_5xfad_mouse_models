### Reformat Metabolome files to BIOM/QZA

```python
metab_files = ['../data/Metabolome/26112024/brain_3xtg_metabolomics_join.csv',
        '../data/Metabolome/26112024/brain_5xfad_metabolomics_join.csv',
        '../data/Metabolome/26112024/liver_3xtg_metabolomics_join.csv',
        '../data/Metabolome/26112024/liver_5xfad_metabolomics_join.csv',
        '../data/Metabolome/26112024/serum_3xtg_metabolomics_join.csv',
        '../data/Metabolome/26112024/serum_5xfad_metabolomics_join.csv',
        '../data/Metabolome/26112024/cecal_3xtg_metabolomics_join.csv',
        '../data/Metabolome/26112024/cecal_5xfad_metabolomics_join.csv',
        '../data/Metabolome/26112024/fecal_gf_long_3xtg_metabolomics_join.csv',
        '../data/Metabolome/26112024/fecal_gf_long_5xfad_metabolomics_join.csv',
        '../data/Metabolome/26112024/fecal_gf_sac_3xtg_metabolomics_join.csv',
        '../data/Metabolome/26112024/fecal_gf_sac_5xfad_metabolomics_join.csv',
        '../data/Metabolome/26112024/fecal_spf_long_3xtg_metabolomics_join.csv',
        '../data/Metabolome/26112024/fecal_spf_long_5xfad_metabolomics_join.csv',
        '../data/Metabolome/26112024/fecal_spf_sac_3xtg_metabolomics_join.csv',
        '../data/Metabolome/26112024/fecal_spf_sac_5xfad_metabolomics_join.csv']
#raw absorbances, no annotations
```

```python
def reformat_metab(file):
    body = file.split("/")[-1]
    borgan = body.split("_")[0]
    gtype = re.search(r'(3xtg|5xfad)', file).group()
    
    if borgan == 'fecal':
        colz = re.search(r'(gf|spf)', file).group()
        phaz = re.search(r'(long|sac)', file).group()
        
        metab = pd.read_csv(file, index_col=0)
        metab = metab.add_prefix('{}_{}_{}_'.format(borgan,colz,gtype))
    
        org_metab = pd.merge(metab, unique_metadata[['study_strain_sex_cage_mouse_tp']], 
                       left_index=True, right_index=True, how='left')
        org_metab.set_index('study_strain_sex_cage_mouse_tp', inplace=True)
    
        org_metab2 = org_metab.T.copy()
        org_metab2 = org_metab2.groupby(by=org_metab2.columns, axis=1).mean()
        org_metab2.index.name='#OTU ID'
    
        org_metab2.to_csv('../data/Metabolome/26112024/rawdata_{}_reformat_{}-{}_mouse_{}.tsv'.format(borgan, colz, gtype, phaz), sep='\t')
        
        fl = 'rawdata_{}_reformat_{}-{}_mouse_{}'.format(borgan, colz, gtype, phaz)
        
        biom_cmd = f"biom convert -i ../data/Metabolome/26112024/{fl}.tsv -o ../data/Metabolome/26112024/{fl}.biom --table-type=\"OTU table\" --to-hdf5"
        qiime_cmd = f"qiime tools import --input-path ../data/Metabolome/26112024/{fl}.biom --input-format BIOMV210Format --type 'FeatureTable[Frequency]' --output-path ../data/Metabolome/26112024/{fl}.qza"

        subprocess.run(biom_cmd, shell=True, check=True)
        subprocess.run(qiime_cmd, shell=True, check=True)
    else:
        metab = pd.read_csv(file, index_col=0)
        metab = metab.add_prefix('{}_'.format(borgan))
    
        org_metab = pd.merge(metab, unique_metadata[['study_strain_sex_cage_mouse_tp']], 
                       left_index=True, right_index=True, how='left')
        org_metab.set_index('study_strain_sex_cage_mouse_tp', inplace=True)
    
        org_metab2 = org_metab.T.copy()
        org_metab2 = org_metab2.groupby(by=org_metab2.columns, axis=1).mean()
        org_metab2.index.name='#OTU ID'
    
        org_metab2.to_csv('../data/Metabolome/26112024/rawdata_{}_reformat_{}_mouse.tsv'.format(borgan, gtype), sep='\t')
    
        fl = 'rawdata_{}_reformat_{}_mouse'.format(borgan, gtype)

        biom_cmd = f"biom convert -i ../data/Metabolome/26112024/{fl}.tsv -o ../data/Metabolome/26112024/{fl}.biom --table-type=\"OTU table\" --to-hdf5"
        qiime_cmd = f"qiime tools import --input-path ../data/Metabolome/26112024/{fl}.biom --input-format BIOMV210Format --type 'FeatureTable[Frequency]' --output-path ../data/Metabolome/26112024/{fl}.qza"

        subprocess.run(biom_cmd, shell=True, check=True)
        subprocess.run(qiime_cmd, shell=True, check=True)
```

```python
for x in metab_files:
    reformat_metab(x)
```

### Reformat Microbiome files to match new sample_ids

```python
micro_3xtg_md = pd.read_csv('../data/Microbiome/3XTG_14748_metadata_overpooling.txt', 
                            sep='\t', index_col=0)
micro_3xtg_md['Study'] = '3xTG'
micro_3xtg_md.rename(columns={'study_type':'study_phase', 
                             'genotype':'mouse_strain'}, inplace=True)
micro_3xtg_md[['that_id1', 'that_id2', 'that_id3', 'that_id4']] = micro_3xtg_md['anonymized_name'].str.split('.', expand=True)
micro_3xtg_md['cage_id'] = micro_3xtg_md.apply(lambda row: row['that_id1'] if str(row['that_id1']).startswith('CAGE') else row['that_id2'] if str(row['that_id2']).startswith('CAGE') else None, axis=1)
micro_3xtg_md['mouse_id'] = micro_3xtg_md.apply(lambda row: row['that_id1'] if str(row['that_id1']).startswith('MOUSE') else 'u', axis=1)
micro_3xtg_md.drop(columns=['that_id1', 'that_id2', 'that_id3', 'that_id4'], inplace=True)
micro_3xtg_md.tail(20)
```

```python
micro_5xfad_md = pd.read_csv('../data/Microbiome/5XFAD_14957_metadata.txt', 
                            sep='\t', index_col=0)
micro_5xfad_md['Study'] = '5xFAD'
micro_5xfad_md['mouse_strain'] = micro_5xfad_md['mouse_strain'].str.split(' ').str[0]
micro_5xfad_md['mouse_strain'] = micro_5xfad_md['mouse_strain'].str.upper()
micro_5xfad_md.rename(columns={'intervention':'study_phase'}, inplace=True)
micro_5xfad_md[['that_id1', 'that_id2', 'that_id3', 'that_id4']] = micro_5xfad_md['anonymized_name'].str.split('.', expand=True)
micro_5xfad_md['cage_id'] = micro_5xfad_md.apply(lambda row: row['that_id1'] if str(row['that_id1']).startswith('CAGE') else row['that_id2'] if str(row['that_id2']).startswith('CAGE') else None, axis=1)
micro_5xfad_md['mouse_id'] = micro_5xfad_md.apply(lambda row: row['that_id1'] if str(row['that_id1']).startswith('MOUSE') else 'u', axis=1)
micro_5xfad_md.drop(columns=['that_id1', 'that_id2', 'that_id3', 'that_id4'], inplace=True)
micro_5xfad_md.tail(20)
```

```python
combo_micro_md = pd.concat([micro_3xtg_md, micro_5xfad_md])
combo_micro_md['Colonized'] = 'SPF'
combo_micro_md.head()
```

```python
combo_micro_md['host_age'] = pd.to_numeric(combo_micro_md['host_age'])
combo_micro_md['study_strain_sex'] = combo_micro_md['Study'].astype(str)+"_"+combo_micro_md['mouse_strain'].astype(str)+"_"+combo_micro_md['sex'].astype(str)
combo_micro_md['study_strain_sex_cage_mouse_tp'] = combo_micro_md['Study'].astype(str)+"_"+combo_micro_md['Colonized'].astype(str)+"-"+combo_micro_md['mouse_strain'].astype(str)+"_"+combo_micro_md['sex'].astype(str)+"_"+combo_micro_md['cage_id'].astype(str)+"-"+combo_micro_md['mouse_id'].astype(str)+"_"+combo_micro_md['host_age'].astype(str)
combo_micro_md['study_strain_sex_cage_mouse_tp'].replace(' ', '', regex=True, inplace=True)
combo_micro_md.tail(10)
```

```python
def reformat_micro(file):
        fle = file.split(".qza")[0]
        micro = pd.read_csv('{}/feature-table.tsv'.format(fle), sep='\t', skiprows=1, index_col=0)
    
        org_micro = pd.merge(micro.T, combo_micro_md[['study_strain_sex_cage_mouse_tp']], 
                       left_index=True, right_index=True, how='left')
        org_micro.set_index('study_strain_sex_cage_mouse_tp', inplace=True)
    
        org_micro2 = org_micro.T.copy()
        org_micro2 = org_micro2.groupby(by=org_micro2.columns, axis=1).mean()
        org_micro2.index.name='#OTU ID'
    
        org_micro2.to_csv('{}/feature-table_reformat.tsv'.format(fle), sep='\t')


        biom_cmd = f"biom convert -i {fle}/feature-table_reformat.tsv -o {fle}_reformat.biom --table-type=\"OTU table\" --to-hdf5"
        qiime_cmd = f"qiime tools import --input-path {fle}_reformat.biom --input-format BIOMV210Format --type 'FeatureTable[Frequency]' --output-path {fle}_reformat.qza"

        subprocess.run(biom_cmd, shell=True, check=True)
        subprocess.run(qiime_cmd, shell=True, check=True)
```

## joint-RPCA (micro-metab)

```python
for pair in [('3XTG_sac_brain-v-micro', brain_3XTG, micro_spf_3XTG_sac), 
              ('3XTG_sac_serum-v-micro', serum_3XTG, micro_spf_3XTG_sac),
             ('3XTG_sac_liver-v-micro', liver_3XTG, micro_spf_3XTG_sac), 
              ('3XTG_sac_cecal-v-micro', cecal_3XTG, micro_spf_3XTG_sac),
             ('3XTG_sac_fecal_metab-v-micro', fecal_spf_3XTG_sac, micro_spf_3XTG_sac),
              ('3XTG_long_fecal_metab-v-micro', fecal_spf_3XTG_long, micro_spf_3XTG_long),
              ('5XFAD_sac_brain-v-micro', brain_5XFAD, micro_spf_5XFAD_sac), 
              ('5XFAD_sac_serum-v-micro', serum_5XFAD, micro_spf_5XFAD_sac),
             ('5XFAD_sac_liver-v-micro', liver_5XFAD, micro_spf_5XFAD_sac), 
              ('5XFAD_sac_cecal-v-micro', cecal_5XFAD, micro_spf_5XFAD_sac),
              ('5XFAD_sac_fecal_metab-v-micro', fecal_spf_5XFAD_sac, micro_spf_5XFAD_sac),
             ('5XFAD_long_fecal_metab-v-micro', fecal_spf_5XFAD_long, micro_spf_5XFAD_long)]:
    
    
    jrpca_results = joint_rpca([pair[1], pair[2]],
                           sample_metadata=metadata,
                           min_feature_frequency=20,
                           max_iterations=4)
    jrpca_results.biplot.save('../data/JointRPCA/{}_biplot-ord.qza'.format(pair[0]))
    jrpca_results.distance_matrix.save('../data/JointRPCA/{}_distance-matrix.qza'.format(pair[0]))
    jrpca_results.cross_validation_error.save('../data/JointRPCA/{}_cross-validation-error.qza'.format(pair[0]))
    
    volatility_plot = volatility(jrpca_results.cross_validation_error.view(q2.Metadata), 'iteration')
    volatility_plot.visualization.save('../data/JointRPCA/{}_volatility_plot.qzv'.format(pair[0]))
    
    df_plt = jrpca_results.cross_validation_error.view(q2.Metadata).to_dataframe()
    plt.errorbar(x=df_plt['iteration'], 
             y=df_plt['mean_CV'], 
             yerr=df_plt['std_CV'],
             c='black')
    plt.ylabel('CV-error', color='black')
    plt.xlabel('iteration', color='black')
    plt.title('{}'.format(pair[0]))
    plt.savefig('../figures/{}_cross-validation-plot.svg'.format(pair[0]), dpi=150)
    plt.show()
    
    emp_plot = biplot(jrpca_results.biplot,
                  metadata,
                  number_of_features=12)
    emp_plot.visualization.save('../data/JointRPCA/{}_emp-biplot.qzv'.format(pair[0]))
    
    permanova_results = beta_group_significance(jrpca_results.distance_matrix, metadata.get_column('strain_sex'))
    permanova_results.visualization.save('../data/JointRPCA/{}_bsig_strain-sex.qzv'.format(pair[0]))
    
    correlation_table = feature_correlation_table(jrpca_results.biplot).correlation_table
    correlation_table.save('../data/JointRPCA/{}_correlation_table.qza'.format(pair[0]))
```

```python
orig_metab_annots = ['../data/Metabolome/26112024/annotation/3xtg_brain_features_info_mn.csv',
               '../data/Metabolome/26112024/annotation/3xtg_serum_features_info_mn.csv',
               '../data/Metabolome/26112024/annotation/3xtg_liver_features_info_mn.csv',
               '../data/Metabolome/26112024/annotation/3xtg_cecum_features_info_mn.csv',
               '../data/Metabolome/26112024/annotation/3xtg_fecal_gf_features_info_mn.csv',
               '../data/Metabolome/26112024/annotation/3xtg_fecal_spf_features_info_mn.csv',
               '../data/Metabolome/26112024/annotation/5xfad_brain_features_info_mn.csv',
               '../data/Metabolome/26112024/annotation/5xfad_serum_features_info_mn.csv',
               '../data/Metabolome/26112024/annotation/5xfad_liver_features_info_mn.csv',
               '../data/Metabolome/26112024/annotation/5xfad_cecum_features_info_mn.csv',
               '../data/Metabolome/26112024/annotation/5xfad_fecal_gf_features_info_mn.csv',
               '../data/Metabolome/26112024/annotation/5xfad_fecal_spf_features_info_mn.csv']
```

```python
def annot_reformat(file):
    body = file.split("/")[-1]
    borgan = body.split("_")[1]
    gtype = re.search(r'(3xtg|5xfad)', file).group()
    
    if borgan == 'fecal':
        colz = re.search(r'(gf|spf)', file).group()
        
        annot = pd.read_csv(file)
        annot['feature_name'] = '{}_{}_{}_'.format(borgan,colz,gtype) + annot['Feature_ID'].astype(str)
        annot.set_index('feature_name', inplace=True)
        annot.to_csv('../data/Metabolome/26112024/annotation/{}_{}_{}_metab-annotation.tsv'.format(borgan, colz, gtype), sep='\t')
    if borgan == 'cecum':
        annot = pd.read_csv(file)
        annot['feature_name'] = 'cecal_'.format(borgan)+annot['Feature_ID'].astype(str)
        annot.set_index('feature_name', inplace=True)
        annot.to_csv('../data/Metabolome/26112024/annotation/{}_{}_metab-annotation.tsv'.format(borgan, gtype), sep='\t')
    else:
        annot = pd.read_csv(file)
        annot['feature_name'] = '{}_'.format(borgan)+annot['Feature_ID'].astype(str)
        annot.set_index('feature_name', inplace=True)
        annot.to_csv('../data/Metabolome/26112024/annotation/{}_{}_metab-annotation.tsv'.format(borgan, gtype), sep='\t')
```

```python
for gtype in ['3XTG', '5XFAD']:
    figsize_factor=1.2

    for organz in ['brain', 'serum', 'liver', 'cecum', 'fecal']:
        if organz == 'fecal':
            metab_3xtg = pd.read_csv('../data/Metabolome/26112024/annotation/{}_spf_3xtg_metab-annotation.tsv'.format(organz),
                                    sep='\t', index_col=0)
            metab_3xtg.dropna(subset=['Compound_Name'], inplace=True)
            metab_5xfad = pd.read_csv('../data/Metabolome/26112024/annotation/{}_spf_5xfad_metab-annotation.tsv'.format(organz),
                                    sep='\t', index_col=0)
            metab_5xfad.dropna(subset=['Compound_Name'], inplace=True)
            both_metab = set(metab_5xfad.Compound_Name.values).intersection(set(metab_3xtg.Compound_Name.values))
            
        else:
            #needs a new one for each body site
            metab_3xtg = pd.read_csv('../data/Metabolome/26112024/annotation/{}_3xtg_metab-annotation.tsv'.format(organz),
                                    sep='\t', index_col=0)
            metab_3xtg.dropna(subset=['Compound_Name'], inplace=True)
            metab_5xfad = pd.read_csv('../data/Metabolome/26112024/annotation/{}_5xfad_metab-annotation.tsv'.format(organz),
                                    sep='\t', index_col=0)
            metab_5xfad.dropna(subset=['Compound_Name'], inplace=True)
            both_metab = set(metab_5xfad.Compound_Name.values).intersection(set(metab_3xtg.Compound_Name.values))
        

        #this works for all sac timepoints
        micro_3xtg = pd.read_csv('../data/Microbiome/annotations/3XTG_fecal_sac_RPCA_Ax1-2_topbot1perc_combo_highprev.tsv', 
                                       sep='\t', index_col=0)
        micro_5xfad = pd.read_csv('../data/Microbiome/annotations/5XFAD_fecal_sac_RPCA_Ax1-2_topbot1perc_combo_highprev.tsv', 
                                       sep='\t', index_col=0)
        both_OGU = set(micro_5xfad.index.values).intersection(set(micro_3xtg.index.values))

        #filter metab and microbiome
        if gtype == '3XTG':
            positive = metab_3xtg[metab_3xtg.Compound_Name.isin(list(both_metab))] 
            #subsets to only those named compound in common between genotypes

            negative = micro_3xtg[micro_3xtg.index.isin(list(both_OGU))].copy()  
            #subsets to only those highly present and differentially abundant (top/bot 1% RPCA) in common between genotypes
            negative['lowest_level'] = negative['Taxon'].str.split(';', 5).str[5]
        if gtype == '5XFAD':
            positive = metab_5xfad[metab_5xfad.Compound_Name.isin(list(both_metab))] 
            #subsets to only those named compound in common between genotypes

            negative = micro_5xfad[micro_5xfad.index.isin(list(both_OGU))].copy()  
            #subsets to only those highly present and differentially abundant (top/bot 1% RPCA) in common between genotypes
            negative['lowest_level'] = negative['Taxon'].str.split(';', 5).str[5]
        
        #load and filter co-ocurrence table
        if organz == 'fecal':
            file = '../data/JointRPCA/{}_sac_{}_metab-v-micro_correlation_table/Correlation.tsv'.format(gtype, organz)
        else:
            file = '../data/JointRPCA/{}_sac_{}-v-micro_correlation_table/Correlation.tsv'.format(gtype, organz)
            
        table = pd.read_csv(file, sep='\t', index_col=0, low_memory=False)
        ctable1 = table.filter(positive.index.values, axis=0)
        ctable2 = ctable1.filter(negative.index.values, axis=1)
        rname1 = pd.merge(ctable2, positive[['Compound_Name']], left_index=True, right_index=True)
        rname1.set_index('Compound_Name', inplace=True)
        rname2 = pd.merge(rname1.T, negative[['lowest_level']], left_index=True, right_index=True)
        rname2.set_index('lowest_level', inplace=True)
        
        if not rname2.empty:
            #clustermap
            num_rows, num_cols = rname2.shape
            base_figsize = (num_cols * 0.8, num_rows * 1)
            figsz = (base_figsize[0] * figsize_factor, base_figsize[1] * figsize_factor)

            g = sns.clustermap(rname2, figsize=figsz, 
                                        yticklabels=True, xticklabels=True,)
            plt.rcParams['svg.fonttype'] = 'none'
            plt.rcParams['font.family'] = 'Arial'
            if organz == 'fecal':
                g.savefig('../figures/VIP-correlations_{}_{}_metab-v-micro_sac.svg'.format(gtype, organz), 
                                dpi=100, bbox_inches='tight')
                plt.close()
            else:
                g.savefig('../figures/VIP-correlations_{}_{}_v_micro_sac.svg'.format(gtype, organz), 
                                dpi=100, bbox_inches='tight')
                plt.close()
        else:
            print("Empty Dataframe - potentially no microbes or metabolites in common")
            print(gtype, organz)
```
